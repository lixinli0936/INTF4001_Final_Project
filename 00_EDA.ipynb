```python
# choose the fields of fact table
facts={}
facts['Accident']=['accident_index','accident_year','longitude','latitude','accident_severity','number_of_vehicles','number_of_casualties','date','time','road_type','weather_conditions']

# fact table name and the referenced dimention table name
dimentions={}
dimentions['Accident']=['accident_severity','road_type','weather_conditions']
```

```python
accidents[facts['Accident']].dtypes
```

```python
casualties=pd.read_csv('./data/Casualty/dft-road-casualty-statistics-casualty-2018.csv',low_memory=False)
casualties.head(10)
```

```python
casualties.info()
```

```python
facts['Casualty']=['accident_index','accident_year','vehicle_reference','casualty_reference','casualty_class','sex_of_casualty','age_of_casualty','age_band_of_casualty','casualty_severity']
dimentions['Casualty']=['casualty_class','sex_of_casualty','age_band_of_casualty','casualty_severity']
```

```python
casualties[facts['Casualty']].dtypes
```

```python
vehicles=pd.read_csv('./data/Vehicle/dft-road-casualty-statistics-vehicle-2018.csv',low_memory=False)
vehicles.head(10)
```

```python
vehicles.info()
```

```python
facts['Vehicle']=['accident_index','accident_year','vehicle_reference','vehicle_type','sex_of_driver','age_of_driver','age_band_of_driver','age_of_vehicle']
dimentions['Vehicle']=['vehicle_type','sex_of_driver','age_band_of_driver']
```

```python
vehicles[facts['Vehicle']].dtypes
```

```python
facts,dimentions
```

```python
field_list = [value for sublist in dimentions.values() for value in sublist]
df=dict.loc[dict['table'].isin(dimentions.keys()) & dict['field name'].isin(field_list)]
df
```

```python
df.groupby('field name')['label'].apply(lambda x: x.str.len().max())
```
